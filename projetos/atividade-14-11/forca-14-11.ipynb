```python
import random
```

```python
def escolher_palavra():
    palavras = ["botafogo", "eduardo", "tiquinho", "segovia", "gatito"]
    return random.choice(palavras)

def exibir_forca(erros):
    partes_corpo = [
        """
           -----
           |   |
               |
               |
               |
               |
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        """
    ]

    return partes_corpo[erros]



def exibir_palavra_oculta(palavra, letras_corretas):
    resultado = ""
    for letra in palavra:
        if letra in letras_corretas:
            resultado += letra + " "
        else:
            resultado += "_ "
    return resultado.strip()



def jogar_forca():
    palavra_secreta = escolher_palavra()
    letras_corretas = []
    tentativas_maximas = len(exibir_forca(0))
    erros = 0

    print("Bem-vindo ao Jogo da Forca!")
    print(exibir_forca(erros))
    print(exibir_palavra_oculta(palavra_secreta, letras_corretas))

    while erros < tentativas_maximas:
        tentativa = input("Digite uma letra: ").lower()

        if tentativa.isalpha() and len(tentativa) == 1:
            if tentativa in letras_corretas:
                print("Você já tentou esta letra. Tente novamente.")
            elif tentativa in palavra_secreta:
                letras_corretas.append(tentativa)
            else:
                erros += 1
            print(exibir_forca(erros))
            print(exibir_palavra_oculta(palavra_secreta, letras_corretas))

            if "_" not in exibir_palavra_oculta(palavra_secreta, letras_corretas):
                print("Parabéns! Você ganhou!")
                break
        else:
            print("Por favor, digite uma única letra válida.")

    if erros == tentativas_maximas:
        print(f"Você perdeu! A palavra correta era '{palavra_secreta}'.")

if __name__ == "__main__":
    jogar_forca()
```
